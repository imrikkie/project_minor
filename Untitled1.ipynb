```python
adj_close = adj_close.ffill().bfill()
adj_close = adj_close.sort_index()
```

```python
returns = adj_close.pct_change().dropna()
```

```python
def classify_state(x, threshold=0.002):
    if x > threshold:
        return 2  # Up
    elif x < -threshold:
        return 0  # Down
    else:
        return 1  # Stable

states = returns.apply(lambda col: col.map(classify_state))

```

```python
print(states)
```

```python
states.columns
```

```python
def transition_matrix(series):
    n = 3
    matrix = np.zeros((n, n))
    for i in range(len(series) - 1):
        current, nxt = series[i], series[i + 1]
        matrix[current][nxt] += 1
    matrix = matrix / matrix.sum(axis=1, keepdims=True)
    return pd.DataFrame(matrix, columns=['Down', 'Stable', 'Up'], index=['Down', 'Stable', 'Up'])


tm_aapl = transition_matrix(states[('Close', 'AAPL')].values)
print(tm_aapl)
```

```python
def markov_predict(stock, adj_close, returns, states, n_days=10):
    s = states[stock].dropna().values
    tm = transition_matrix(s)


    up_mean = returns[stock][states[stock] == 2].mean()
    down_mean = returns[stock][states[stock] == 0].mean()
    stable_mean = returns[stock][states[stock] == 1].mean()


    last_price = adj_close[stock].iloc[-1]
    last_state = s[-1]
    current_state = np.zeros(3)
    current_state[last_state] = 1

    predicted_prices = []
    current_price = last_price
    current_prob = current_state

    for i in range(n_days):
        current_prob = np.dot(current_prob, tm)
        expected_return = (
            current_prob[2] * up_mean +
            current_prob[1] * stable_mean +
            current_prob[0] * down_mean
        )
        current_price = current_price * (1 + expected_return)
        predicted_prices.append(current_price)

    return predicted_prices
```

```python
n_days = 10
predicted_df = pd.DataFrame()

for stock in adj_close.columns:
    predicted_df[stock] = markov_predict(stock, adj_close, returns, states, n_days)

predicted_df.index = pd.RangeIndex(1, n_days + 1, name='Day')
print("Predicted future prices for all stocks:")
print(predicted_df.head())
```

```python
def align_series(pred, actual):
    min_len = min(len(pred), len(actual))
    return pred.iloc[:min_len], actual.iloc[:min_len]
```

```python
compare_df = pd.DataFrame()

for stock in predicted_df.columns:
    pred, act = align_series(predicted_df[stock], actual_df[stock])
    compare_df[(stock, "Predicted")] = pred.values
    compare_df[(stock, "Actual")] = act.values
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

for stock in predicted_df.columns:

    pred = predicted_df[stock].values
    act = actual_df[stock].values

    # Align lengths
    min_len = min(len(pred), len(act))
    pred = pred[:min_len]
    act = act[:min_len]

    # Remove rows where either is NaN
    mask = ~np.isnan(pred) & ~np.isnan(act)
    pred_clean = pred[mask]
    act_clean = act[mask]

    if len(pred_clean) == 0:
        print(stock, " → All values are NaN. Skipping.")
        continue

    mae = mean_absolute_error(act_clean, pred_clean)
    mse = mean_squared_error(act_clean, pred_clean)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((act_clean - pred_clean) / act_clean)) * 100

    print(stock)
    print("MAE:", mae)
    print("RMSE:", rmse)
    print("MAPE:", mape)
    print("---------------------------")


```
