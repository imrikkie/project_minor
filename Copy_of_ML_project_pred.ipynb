```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
tickers = [
    'NVDA', 'ITC', 'RELIANCE.NS', 'WDAY', 'TATAMOTORS.NS',
    'AAPL', 'ZS', 'BE', 'JNJ', 'SUNPHARMA.NS'
]

data = yf.download(tickers, start='2015-10-01', end='2025-10-01', auto_adjust=True)
close_prices = data['Close']
```

```python
def create_features(df, n_lag=5):
    X, y = [], []
    for i in range(n_lag, len(df)):
        X.append(df.iloc[i-n_lag:i].values)
        y.append(df.iloc[i])
    return np.array(X), np.array(y)
```

```python
results = {}
future_predictions = {}

for ticker in tickers:
    print(f"\nTraining models for {ticker}...")

    series = close_prices[ticker].dropna()

    X, y = create_features(series)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, shuffle=False
    )

    # Reshape for ML models
    X_train_r = X_train.reshape(len(X_train), -1)
    X_test_r = X_test.reshape(len(X_test), -1)

    models = {
        "Linear Regression": LinearRegression(),
        "Random Forest": RandomForestRegressor(n_estimators=250, max_depth=10),
        "XGBoost": XGBRegressor(n_estimators=300, learning_rate=0.05, max_depth=5)
    }

    model_scores = {}

    # Train all models
    for name, model in models.items():
        model.fit(X_train_r, y_train)
        pred = model.predict(X_test_r)
        mae = mean_absolute_error(y_test, pred)
        model_scores[name] = mae

    # Save model performance
    results[ticker] = model_scores

    # Best performing model for this ticker
    best_model_name = min(model_scores, key=model_scores.get)
    best_model = models[best_model_name]

    # Predict next 5 future days
    last_vals = series.values[-5:]

    preds = []
    cur = last_vals.copy()

    for _ in range(5):
        inp = cur.reshape(1, -1)
        next_pred = best_model.predict(inp)[0]
        preds.append(next_pred)
        cur = np.append(cur[1:], next_pred)

    future_predictions[ticker] = preds
```

```python
df_pred = pd.DataFrame(future_predictions, index=[1,2,3,4,5])
print("\nPredicted future prices for all stocks:\n")
print(df_pred)
```

```python
overall = {}
for t in results:
    for m in results[t]:
        overall.setdefault(m, 0)
        overall[m] += results[t][m]

best_overall = min(overall, key=overall.get)
print("\nOverall best model:", best_overall)
```
