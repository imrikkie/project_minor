```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from math import sqrt
```

```python
#1. SELECT TICKER & PREPARE PRICE SERIES

# If multi-ticker: data has MultiIndex columns like ('Close','INFY.NS')
if isinstance(data.columns, pd.MultiIndex):
    ticker = 'INFY.NS'       # change ticker here
    close = data['Close'][ticker].dropna()
else:
    close = data['Close'].dropna()

df = close.to_frame(name='Close')
```

```python
#2. FEATURE ENGINEERING
# returns
df['return_1'] = df['Close'].pct_change(1)
df['return_5'] = df['Close'].pct_change(5)

# lag features
for lag in [1,2,3,5,7]:
    df[f'lag_{lag}'] = df['Close'].shift(lag)

# rolling averages
df['rmean_7'] = df['Close'].rolling(7).mean()
df['rmean_21'] = df['Close'].rolling(21).mean()

# rolling std (volatility)
df['rstd_21'] = df['Close'].rolling(21).std()

# momentum
df['momentum_7'] = df['Close'] - df['Close'].shift(7)

# RSI (simple implementation)
delta = df['Close'].diff()
up = delta.clip(lower=0)
down = -1 * delta.clip(upper=0)
roll_up = up.ewm(span=14).mean()
roll_down = down.ewm(span=14).mean()
rs = roll_up / roll_down
df['rsi_14'] = 100 - (100 / (1 + rs))

# MACD
ema12 = df['Close'].ewm(span=12).mean()
ema26 = df['Close'].ewm(span=26).mean()
df['macd'] = ema12 - ema26
df['macd_signal'] = df['macd'].ewm(span=9).mean()

# Target → next day Close
df['target'] = df['Close'].shift(-1)

# remove NaNs
df = df.dropna()

# ---------------------------------------------------
# 3. TRAIN-TEST SPLIT
# ---------------------------------------------------

X = df.drop(columns=['target'])
y = df['target']

test_size = int(len(df) * 0.2)
X_train, X_test = X.iloc[:-test_size], X.iloc[-test_size:]
y_train, y_test = y.iloc[:-test_size], y.iloc[-test_size:]
```

```python
# 4. XGBOOST MODEL (pipeline with scaler)

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("xgb", xgb.XGBRegressor(
        objective='reg:squarederror',
        n_estimators=500,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    ))
])

pipeline.fit(X_train, y_train)
```

```python
# 5. EVALUATION
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# y_test and y_pred already defined
y_pred = pipeline.predict(X_test)

rmse = sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print("RMSE:", rmse)
print("MAE :", mae)
print("R²  :", r2)
```
